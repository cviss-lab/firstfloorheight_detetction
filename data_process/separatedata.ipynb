```python
import os
import shutil

def divide_images_into_sections(folder_path, num_sections):
    # Get a list of all image files in the folder
    image_files = [f for f in os.listdir(folder_path) if os.path.isfile(os.path.join(folder_path, f))]

    # Calculate the number of images in each section
    num_images = len(image_files)
    print(num_images)
    images_per_section = num_images // num_sections
    print(images_per_section)

    
    # Create subfolders to store the divided images
    for section in range(1, num_sections + 1):
        section_folder = os.path.join(folder_path, f'Section_{section}')
        os.makedirs(section_folder, exist_ok=True)

    # Move images into their corresponding sections
    for idx, image_file in enumerate(image_files):
        source_path = os.path.join(folder_path, image_file)
        section = (idx // images_per_section) + 1
        destination_path = os.path.join(folder_path, f'Section_{section}', image_file)
        shutil.move(source_path, destination_path)
    

if __name__ == "__main__":
    folder_path = "0830undistort"
    num_sections = 10

    divide_images_into_sections(folder_path, num_sections)
```
